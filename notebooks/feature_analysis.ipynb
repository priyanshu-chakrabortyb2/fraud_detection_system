```python
import pandas as pd
train_df=pd.read_csv("../data/processed/train_df.csv")
```

```python
corr = train_df.corr(numeric_only=True)

corr["isFraud"].sort_values(ascending=False)
```

```python
fraud_corr = (corr["isFraud"].sort_values(ascending=False))

print(fraud_corr)
```

```python
feature_corr = (
    train_df.corr(numeric_only=True)["isFraud"]
    .sort_values(ascending=False)
)

print(feature_corr)
```

```python
import matplotlib.pyplot as plt

feature_corr.drop("isFraud").sort_values().plot(kind="barh",figsize=(10,6))

plt.title("Feature Correlation with Fraud")
plt.show()
```

```python
engineered_features=[
    "balance_diff_orig",
    "balance_diff_dest",
    "amount_to_balance_ratio",
    "is_zero_balance_org",
    "is_zero_balance_dest",
    "hour_of_day"
]

print(
    train_df[engineered_features + ["isFraud"]]
    .corr(numeric_only=True)["isFraud"]
    .sort_values(ascending=False)
)
```
